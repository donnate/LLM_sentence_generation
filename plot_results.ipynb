```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd




##### joad json data
# Load the JSON data


results42 = pd.read_csv('/Users/clairedonnat/Documents/LLM_topics/results42.csv')
results43 = pd.read_csv('/Users/clairedonnat/Documents/LLM_topics/results43.csv')
results45 = pd.read_csv('/Users/clairedonnat/Documents/LLM_topics/results45.csv')
results46 = pd.read_csv('/Users/clairedonnat/Documents/LLM_topics/results46.csv')
results47 = pd.read_csv('/Users/clairedonnat/Documents/LLM_topics/results47.csv')
```

```python
res = pd.concat([results42, results43, results45, results46, results47], axis=0)
```

```python
res
```

```python
res_plsi = res.groupby(['k']).agg({'acc_pLSI': ['mean', 'median', 'std']}).reset_index()
res_lda = res.groupby(['k']).agg({'acc_LDA': ['mean', 'median', 'std']}).reset_index()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))

# Plot pLSI
plt.errorbar(res_plsi['k'], res_plsi['acc_pLSI']['mean'], yerr=res_plsi['acc_pLSI']['std'],
             label='pLSI', fmt='-o', capsize=3)

# Plot LDA
plt.errorbar(res_lda['k'], res_lda['acc_LDA']['mean'], yerr=res_lda['acc_LDA']['std'],
             label='LDA', fmt='-s', capsize=3)

plt.xlabel("Number of alternatives used (k)")
plt.ylabel("Accuracy")
plt.title("Accuracy vs. k for LDA and pLSI")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
